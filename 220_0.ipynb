```python
import numpy as np
arr = np.array([10, 20, 30, 40, 50])
print(arr)
a = np.zeros(len(arr), dtype=int)
i = len(arr)-1
count = 0
while i>=0:
    print(arr[i])
    a[count] = arr[i]
    i -= 1
    count += 1
print(a)
```

```python
# right shift
arr = np.array([10, 20, 30, 40, 50])
n = int(input())
count = 0
while count<n:
    i = len(arr)-1
    while i>0:
        arr[i] = arr[i-1]
        i-=1
    arr[i] = 0
    count+=1
    print(arr)
```

```python
# left shift
arr = np.array([10, 20, 30, 40, 50])
n = int(input())
count = 0
while count<n:
    i = 0
    while i<len(arr)-1:
        arr[i] = arr[i+1]
        i+=1
    arr[i] = 0
    count+=1
    print(arr)
```

```python
# right rotate
arr = np.array([10, 20, 30, 40, 50])
n = int(input())
count = 0
while count<n:
    i = len(arr)-1
    x = arr[i]
    while i>0:
        arr[i] = arr[i-1]
        i-=1
    arr[i] = x
    count+=1
    print(arr)
```

```python
# left rotate
arr = np.array([10, 20, 30, 40, 50])
n = int(input())
count = 0
while count<n:
    i = 0
    x = arr[i]
    while i<len(arr)-1:
        arr[i] = arr[i+1]
        i+=1
    arr[i] = x
    count+=1
    print(arr)
```
