```python
class BTNode:
  def __init__(self, elem):
    self.elem = elem
    self.right = None
    self.left = None
```

```python
def inorder(root):
  if root == None:
    return

  inorder(root.left)
  print(root.elem, end = ' ')
  inorder(root.right)
```

```python
def tree_construction(arr, i = 1):
  if i>=len(arr) or arr[i] == None:
    return None
  p = BTNode(arr[i])
  p.left = tree_construction(arr, 2*i)
  p.right = tree_construction(arr, 2*i+1)
  return p


root2 = tree_construction([None, 'A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', None, None, None, 'I', 'J', None, 'k'])
inorder(root2)
```

```python
def LCA(root, x, y):
  if root == None:
    return None
  elif root.elem == x and root.elem == y:
    return root.elem
  elif root.elem < x and root.elem < y:
    return LCA(root.right, x, y).elem
  elif root.elem > x and root.elem > y:
    return LCA(root.left, x, y).elem
  else:
    return root


#DRIVER CODE
#Write by yourself from the given tree (Create parent node and its corresponding left and right children nodes)
#check all the sample inputs given
#You can take help by seeing the driver code of Lab-6

root = tree_construction([None, 15, 10, 25, 8, 12, 20, 30, 6, 9, None, None, 18, 22, None, None])
print(LCA(root, 6, 12))
```

```python
def find_path(root, key):
  if root == None:
    return 'No Path Found'
  elif key == root.elem:
    return [root.elem]
  elif key < root.elem:
    path1 = find_path(root.left, key)
    if path1 != 'No Path Found':
      return [root.elem] + path1
  else:
    path2 = find_path(root.right, key)
    if path2 != 'No Path Found':
      return [root.elem] + path2
  return 'No Path Found'

#DRIVER CODE
#Write by yourself from the given tree
root = tree_construction([None, 30, 10, 40, 3, 15, 35, 55])
print(find_path(root,15))
#This should print [30,10,15]

print(find_path(root,50))
#This should print No Path Found
```

```python
def sum_of_leaves(root, sum):
  if root is None:
    return sum
  if root.left is None and root.right is None:
    sum += root.elem
    return sum
  sum = sum_of_leaves(root.left, sum)
  sum = sum_of_leaves(root.right, sum)
  return sum


#DRIVER CODE
#Write by yourself from the given tree
root = tree_construction([None, 30, 10, 40, 3, 15, 35, 55, 2, None, None, None, None, 36, None, None])
print(sum_of_leaves(root, 0))
```
