**Assignment Part 1: Doubly Linked List**

For Assignment Part 1, you should write your full code in the following cells along with your driver codes on your own.

```python
#Assignment Part 1

class Patient:
  #write a constructor
  def __init__(self, id, name, age, bloodgroup, next=None, prev=None):
    self.id = id
    self.name = name
    self.age = age
    self.bloodgroup = bloodgroup
    self.next = next
    self.prev = prev
```

```python
class WRM:

  def __init__(self):
    #Creating the dummy head
    self.dh = Patient(None,None,None,None)
    self.dh.next = self.dh
    self.dh.prev = self.dh

  def RegisterPatient(self,id, name, age, bloodgroup):
    new = Patient(id, name, age, bloodgroup)
    new.next = self.dh
    new.prev = self.dh.prev
    self.dh.prev.next = new
    self.dh.prev = new

  def ServePatient(self):
    self.dh.next = self.dh.next.next
    self.dh.next.prev = self.dh

  def ShowAllPatient(self):
    curr = self.dh.next
    if curr == self.dh:
      print('No patient in the waiting room')
    else:
      while curr != self.dh:
        print(curr.name)
        curr = curr.next

  def CanDoctorGoHome(self):
    if self.dh.next == self.dh:
      return True
    return False

  def CancelAll(self):
    self.dh.next = self.dh
    self.dh.prev = self.dh

  def ReverseTheLine(self):
    curr = self.dh.next
    while curr != self.dh:
      curr.next, curr.prev = curr.prev, curr.next
      curr = curr.prev
    self.dh.next, self.dh.prev = self.dh.prev, self.dh.next
```

```python
#Write a Tester Code in this cell
print("**Welcome to Waiting Room Management System**")
def main():
  wrm = WRM()
  while True:
      print("\nOptions:")
      print("1. Add Patient")
      print("2. Serve Patient")
      print("3. Show All Patients")
      print("4. Can Doctor go Home?")
      print("5. Cancel all Appointments")
      print("6. Reverse the Line")
      print("0. Exit")

      option = int(input("Select an option: "))

      if option == 1:
          id = int(input("Enter ID: "))
          name = input("Enter Name: ")
          age = int(input("Enter Age: "))
          bloodgroup = input("Enter Blood Group: ")
          wrm.RegisterPatient(id, name, age, bloodgroup)
      elif option == 2:
          wrm.ServePatient()
      elif option == 3:
          wrm.ShowAllPatient()
      elif option == 4:
          print(wrm.CanDoctorGoHome())
      elif option == 5:
          wrm.CancelAll()
      elif option == 6:
          wrm.ReverseTheLine()
      elif option == 0:
          break
      else:
          print("Invalid option, please try again.")

if __name__ == "__main__":
  main()
```

**Assignment Part 2: Stack**

Linked List based Stack is implemented in the following cell.

```python
class Node:
  def __init__(self,elem=None,next=None):
    self.elem = elem
    self.next = next

class Stack:
  def __init__(self):
    self.__top = None

  def push(self,elem):
    nn = Node(elem,self.__top)
    self.__top = nn

  def pop(self):
    if self.__top == None:
      #print('Stack Underflow')
      return None
    e = self.__top
    self.__top = self.__top.next
    return e.elem

  def peek(self):
    if self.__top == None:
      #print('Stack Underflow')
      return None
    return self.__top.elem

  def isEmpty(self):
    return self.__top == None
```

```python
#You can run this driver code cell to understand the methods of Stack class
st = Stack()
st.push(4)
st.push(3)
st.push(5)
st.push(1)
st.push(9)

print('Peeked Element: ',st.peek())
print('Popped Element: ',st.pop())
print('Popped Element: ',st.pop())
print('Popped Element: ',st.pop())
print('Peeked Element: ',st.peek())
print('Popped Element: ',st.pop())
print('Popped Element: ',st.pop())
print('Peeked Element: ',st.peek())
print('Popped Element: ',st.pop())
print(st.isEmpty())
```

You can print your stack using this code segment

```python
def print_stack(st):
  if st.isEmpty():
    return
  p = st.pop()
  print('|',p,end=' ')
  if p<10:
    print('|')
  else:
    print('|')
  #print('------')
  print_stack(st)
  st.push(p)

st = Stack()
st.push(4)
st.push(3)
st.push(5)
st.push(1)
st.push(9)
print_stack(st)
print('------')
```

Task 1: Diamond Count

```python
def diamond_count(stack,string):
  count = 0
  for i in string:
    if i == '<':
      stack.push(i)
    elif i == '>' and stack.isEmpty() == False:
      stack.pop()
      count += 1
  return count

print('Test 01')
stack = Stack()
string = '<..><.<..>> '
returned_value = diamond_count(stack,string)
print(f'Number of Diamonds: {returned_value}') #This should print 3
unittest.output_test(returned_value, 3)
print('-----------------------------------------')


print('Test 02')
stack = Stack()
string = '<<<..<......<<<<....>'
returned_value = diamond_count(stack,string)
print(f'Number of Diamonds: {returned_value}') #This should print 1
unittest.output_test(returned_value, 1)
print('-----------------------------------------')


print('Test 03')
stack = Stack()
string = '>>><...<<..>>...>...>>>'
returned_value = diamond_count(stack,string)
print(f'Number of Diamonds: {returned_value}') #This should print 3
unittest.output_test(returned_value, 3)
print('-----------------------------------------')
```

Task 2: Tower of Blocks

```python
def remove_block(stack, n):
  stack = Stack()
  for i in range(n):
    stack.push(st.pop())
  stack.pop()
  while stack.isEmpty() == False:
    st.push(stack.pop())
  return st

print('Test 01')
st = Stack()
st.push(4)
st.push(19)
st.push(23)
st.push(17)
st.push(5)
print('Stack:')
print_stack(st)
print('------')
remove_block(st,2)
print('After Removal')
print_stack(st)
print('------')

print()
print('======================================')
print()

print('Test 02')
st = Stack()
st.push(73)
st.push(85)
st.push(15)
st.push(41)
print('Stack:')
print_stack(st)
print('------')
remove_block(st,3)
print('After Removal')
print_stack(st)
print('------')

print()
print('======================================')
print()
```

Task 3: Stack Reverse

```python
def conditional_reverse(stack):
  stack = Stack()
  stack.push(st.pop())
  while st.isEmpty() == False:
    if st.peek() == stack.peek():
      st.pop()
    else:
      stack.push(st.pop())
  return stack

print('Test 01')
st=Stack()
st.push(10)
st.push(10)
st.push(20)
st.push(20)
st.push(30)
st.push(10)
st.push(50)
print('Stack:')
print_stack(st)
print('------')
reversed_stack=conditional_reverse(st)
print('Conditional Reversed Stack:')
print_stack(reversed_stack) # This stack contains 50, 10, 30, 20, 10 in this order whereas top element should be 10
print('------')
```
