```python
class Node:
    def __init__(self, vertex , weight = 1):
        self.vertex = vertex
        self.weight = weight
        self.next = None
class LinkedList:
    def __init__(self):
        self.head = None
    def addedge(self,vertex ,weight = 1):
        newNode = Node(vertex , weight)
        newNode.next = self.head
        self.head = newNode

class Graph:
    def __init__(self, vertices):
        self.vertices = vertices
        self.adj_list = {}
        for i in vertices:
            self.adj_list[i] = LinkedList()
    def add_edge(self, vertex , dest , weight = 1 ):
        self.adj_list[vertex].addedge(dest , weight)
    def max_degree_count(self):
        mmax_degree = 0
        max_vertex = None
        for i in self.vertices:
            degree = 0
            temp = self.adj_list[i].head
            while temp != None:
                degree += 1
                temp = temp.next
            if degree > mmax_degree:
                mmax_degree = degree
                max_vertex = i
        return max_vertex , mmax_degree
    def max_weight_count(self):
        mmax_weight = 0
        max_vertex = None
        for i in self.vertices:
            weight = 0
            temp = self.adj_list[i].head
            while temp != None:
                weight += temp.weight
                temp = temp.next
            if weight > mmax_weight:
                mmax_weight = weight
                max_vertex = i
        return max_vertex , mmax_weight
    def directed_to_undirected(self):
        for vertex in self.vertices:
            temp = self.adj_list[vertex].head
            while temp is not None:
                reverse_exists = False
                reverse_temp = self.adj_list[temp.vertex].head
                while reverse_temp is not None:
                    if reverse_temp.vertex == vertex:
                        reverse_exists = True
                        break
                    reverse_temp = reverse_temp.next
                if not reverse_exists:
                    self.add_edge(temp.vertex, vertex, temp.weight)
                temp = temp.next
    def print_graph(self):
        for vertex in self.vertices:
            print(vertex, end=" -> ")
            temp = self.adj_list[vertex].head
            while temp is not None:
                print(f"({temp.vertex}, {temp.weight})", end=" -> ")
                temp = temp.next
            print("None")
```
